## 4. Distributions and Outliers

### A. Plotting numerical columns' distributions

```python
df.hist(bins=30, figsize=(15, 15))
plt.suptitle('Distribution of Numeric Features')
plt.savefig('../outputs/notebook 1/numeric_columns_distribution.png')
plt.show()
plt.close()
```

### B. Selecting right-skewed columns for log transformation

```python
skewed_cols = ['Price', 'lot_area_renov', 'living_area_renov', 'Area of the basement', 'Area of the house(excluding basement)', 'lot area', 'living area']
```

### C. Checking for outliers before transformations to compare with aftermath

```python
for c in skewed_cols:
    plt.figure(figsize=(3,3))
    df[c].plot.box()
    plt.show()
```

### D. Log-Transformation of Dataframe

```python
'''
A copy of dataframe will be used because 
we will use a Pipeline with TransformTargetRegressor to automate 
streamlit app's user input transformations and prediction inverse_transformation
'''
df_copy = df.copy()
df_copy[skewed_cols] = np.log1p(df_copy[skewed_cols])
```

### E. Distributions after transformations

```python
df_copy.hist(bins=30, figsize=(15, 15))
plt.suptitle('Distribution of Numeric Features')
plt.savefig('../outputs/notebook 1/transformed_columns_distribution.png')
plt.show()
plt.close()
```

### F. Removing outliers using IQR method

```python
initial_count = len(df_copy) # rows before removing them

for c in skewed_cols:
    # Calculate Q1 (25th percentile) and Q3 (75th percentile)
    Q1 = df_copy[c].quantile(0.25)
    Q3 = df_copy[c].quantile(0.75)
    IQR = Q3 - Q1
    
    # Defining bounds
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    df_copy = df_copy[(df_copy[c] >= lower_bound) & (df_copy[c] <= upper_bound)]

dropped_count = initial_count - len(df_copy)
print(f"Rows before: {initial_count}")
print(f"Rows dropped: {dropped_count} ({(dropped_count/initial_count)*100:.2f}%)")
print(f"Rows after:  {len(df_copy)}")
```

### G. Outliers plotting after transformations and removal

```python
for c in skewed_cols:
    plt.figure(figsize=(3,3))
    df_copy[c].plot.box()
    plt.show()
```

We can see there are still some points just outside of box-plots in some columns. At this point, these perhaps have genuine data telling extremmely affordable and luxury houses. So we will retain them in data, and if our model doesn't perform well, we can remove them then.

## 5. Correlation between Variables

```python
# Plotting pearson correlation as it directly measures linear strength between variables
corr = df_copy.corr()
mask = np.triu(np.ones_like(corr, dtype=bool))

plt.figure(figsize=(20,6))
sns.heatmap(corr, mask=mask, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Pearson Correlation Heatmap')

'''
Plotting spearman correlation to measure multicollinearity between variables because
it captures non-linear monotonic relationships and is more robust to outliers, which helps in identifying redundant features
'''
s_corr = df_copy.corr(method='spearman')

s_mask = np.triu(np.ones_like(corr, dtype=bool))

plt.figure(figsize=(20,6))
sns.heatmap(s_corr, mask=s_mask, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Spearman Correlation Heatmap')

plt.show()
```
